##  funcation	to	re	calculate	the	null	values	and	assign	them	into	null	values ##

```python
mask = df['OpeningQuantity'].isna() & df['ReceiptQuantity'].notna() & df['IntransitQuantity'].notna() & df['SalesQuantity'].notna() & df['ClosingQuantity'].notna()

df.loc[mask, 'OpeningQuantity'] = (
    df.loc[mask, 'ClosingQuantity']
    - df.loc[mask, 'ReceiptQuantity']
    + df.loc[mask, 'IntransitQuantity']
    - df.loc[mask, 'SalesQuantity']
).astype(int)
```

```python
mask = df['ReceiptQuantity'].isna() & df['OpeningQuantity'].notna() & df['IntransitQuantity'].notna() & df['SalesQuantity'].notna() & df['ClosingQuantity'].notna()

df.loc[mask, 'ReceiptQuantity'] = (
    df.loc[mask, 'ClosingQuantity']
    - df.loc[mask, 'OpeningQuantity']
    - df.loc[mask, 'IntransitQuantity']
    + df.loc[mask, 'SalesQuantity']
).astype(float)
```

```python
mask = df['IntransitQuantity'].isna() & df['OpeningQuantity'].notna() & df['ReceiptQuantity'].notna() & df['SalesQuantity'].notna() & df['ClosingQuantity'].notna()

df.loc[mask, 'IntransitQuantity'] = (
    df.loc[mask, 'ClosingQuantity']
    - df.loc[mask, 'OpeningQuantity']
    - df.loc[mask, 'ReceiptQuantity']
    + df.loc[mask, 'SalesQuantity']
).astype(int)
```

```python
mask = df['SalesQuantity'].isna() & df['OpeningQuantity'].notna() & df['ReceiptQuantity'].notna() & df['IntransitQuantity'].notna() & df['ClosingQuantity'].notna()

df.loc[mask, 'SalesQuantity'] = (
    df.loc[mask, 'ClosingQuantity']
    - df.loc[mask, 'OpeningQuantity']
    - df.loc[mask, 'ReceiptQuantity']
    + df.loc[mask, 'IntransitQuantity']
).astype(int)
```

```python
mask = df['ClosingQuantity'].isna() & df['OpeningQuantity'].notna() & df['ReceiptQuantity'].notna() & df['IntransitQuantity'].notna() & df['SalesQuantity'].notna()

df.loc[mask, 'ClosingQuantity'] = (
    df.loc[mask, 'OpeningQuantity']
    + df.loc[mask, 'ReceiptQuantity']
    - df.loc[mask, 'IntransitQuantity']
    - df.loc[mask, 'SalesQuantity']
).astype(int)
```
